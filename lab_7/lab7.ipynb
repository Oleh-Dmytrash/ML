```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import Layout

colorRadio = widgets.Dropdown(
    options=[('Green', 'g'), ('Red', 'r'), ('Yellow', 'y'), ('Blue', 'b')],
    description='Color',
    disabled=False
)

display(colorRadio)
```

```python
markerRadio = widgets.Dropdown(
    options=[('Circle', 'o'), ('Triangle', '^'), ('Line', '-'), ('Dashed line', '--')],
    description='Marker style',
    disabled=False
)

display(markerRadio)
```

```python
markerWidth = widgets.FloatSlider(
    value=0.5,
    min=0,
    max=10.0,
    step=0.05,
    description='Marker width:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

display(markerWidth)
```

```python
pointsNumber = widgets.IntSlider(
    value=1000,
    min=1,
    max=5000,
    step=1,
    description='Points number:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
)

display(pointsNumber)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 5 * np.sin(10 * x) * np.sin(3 * x) / (x ** (1/2))

x = np.linspace(1, 7, int(pointsNumber.value))
y = f(x)

style = colorRadio.value + markerRadio.value
plt.plot(x, y, style, linewidth = markerWidth.value, color = colorRadio.value)

plt.legend(["5*sin(10*x)*sin(3*x)/x^(1/2)"], loc='lower right')
plt.xlabel("x")
plt.ylabel('y')

plt.show()
```

```python
from sklearn.datasets import fetch_lfw_people

dataset = dataset =  fetch_lfw_people()
x = dataset.data[:1000]
```

```python
firstColumn = x[:,0]
firstColumn = firstColumn[firstColumn != 0]

plt.hist(firstColumn, 10, color="red")
plt.xlabel("first column value")
plt.ylabel("frequency")

plt.savefig("first.png", dpi=200)
plt.show()
```

```python
secondColumn = x[:,1]
secondColumn = secondColumn[secondColumn != 0]

plt.hist(secondColumn, 10, color="red")
plt.xlabel("second column value")
plt.ylabel("frequency")

plt.savefig("second.png", dpi=200)
plt.show()
```

```python
someColumn = x[:,-87]
someColumn = someColumn[someColumn != 0]

plt.hist(someColumn, 10, color="red")
plt.xlabel("some column value")
plt.ylabel("frequency")

plt.savefig("some.png", dpi=200)
plt.show()
```

```python
plt.plot(firstColumn, "r-",
         secondColumn, "r--",
         someColumn, 'r:')

plt.legend(["first column", "second column", "some column"], loc="upper left")

plt.savefig("plot.png", dpi=200)
plt.show()
```
