```python
import nltk
from nltk import *
from nltk import FreqDist, ConditionalFreqDist
from nltk.corpus import gutenberg, brown

genres = ['adventure', 'government', 'hobbies', 'humor', 'news', 'religion', 'science_fiction']
modals = ['can', 'could', 'may', 'might', 'must', 'will', 'need']
prepositions = ['of', 'on', 'at', 'in', 'from']

cfd = ConditionalFreqDist(
       (genre, word.lower())
       for genre in brown.categories()
       for word in brown.words(categories=genre)
)

cfd.plot(conditions=genres, samples=modals)
cfd.plot(conditions=genres, samples=prepositions)
```

```python
import nltk
from nltk import *
from nltk.corpus import brown

for genre in brown.categories():
    words = brown.words(categories=genre)
    wordsCount = len(words)
    uniqueWordsCount = len(set(words))
    result = wordsCount / uniqueWordsCount
    
    print("%15s %10f" % (genre, result))
```
