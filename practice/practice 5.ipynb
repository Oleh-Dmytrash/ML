```python
from nltk.corpus import gutenberg
from collections import Counter

words = [word.lower() for word in gutenberg.words('whitman-leaves.txt')]
words_5more = [word for word in words if len(word) > 5]

counter = Counter(words_5more)
vocab = counter.most_common(25)
for word, count in vocab:
    print(f'"{word}": {count}')
```

```python
from nltk.corpus import gutenberg
from nltk import FreqDist

words = [word.lower() for word in gutenberg.words('whitman-leaves.txt')]

fdist = FreqDist(words)

words_5more = [word for word in words if fdist[word] >= 5]
print(set(words_5more))
```

```python
from nltk import ConditionalFreqDist
from nltk.corpus import brown

genres = ['news', 'romance']
days = ['sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday']

cfd = ConditionalFreqDist((genre, word.lower())
    for genre in genres for word in brown.words(categories=genre)
        if word.lower() in days)
cfd.tabulate()
cfd.plot()
```
